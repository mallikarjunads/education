```python
def malli(plot,title,xlabel,ylabel):
    plot.set_title(title,fontsize=13,weight="bold")
    plot.set_xlabel(xlabel,fontsize=13,weight="bold")
    plot.set_ylabel(ylabel,fontsize=13,weight="bold")
```

```python
d=rank["country"].value_counts().to_frame().head(20)
```

```python
d
```

```python
plt.figure(figsize=(15,6))
sns.set_style("darkgrid")
ax = sns.barplot(data=d,x="country",y="count",palette="husl")
malli(ax,"top 20 countries got most ranks","countries","count")
for i in ax.containers:
    ax.bar_label(i)
ax.tick_params("x",rotation=90)
```

```python
country = rank.country.value_counts().head(20)
plt.figure(figsize=(8,5))
plt.title("Top 20 countries with most rank univerisities",fontdict={'fontweight':'bold','fontsize':12})
plt.barh(country.index,country.values)
plt.show()
```

```python
sorted_df =rank.sort_values(by=['alumni_employment'],ascending=False).head(10) #sorting the dataframe based on alumni_employment
plt.figure(figsize=(12,8))

plt.title("Top 10 Univerity with highest alumini_employment",fontdict={'fontweight':'bold','fontsize':12})

plt.barh(sorted_df.institution,sorted_df.alumni_employment)
for i,v in enumerate(sorted_df.alumni_employment):
    plt.text(v,i,str(v))
plt.show()
```
